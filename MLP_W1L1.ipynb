# data is normaly distributed, what are outliers?
- Far away from mean
- empirical rule
- $\mu - $ to $\mu + $ = 68 %
- $\mu - 2$ to $\mu + 2$ = 95 %
- $\mu - 3$ to $\mu + 3$ = 99.7 %
